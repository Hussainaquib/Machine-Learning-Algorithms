```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv('kmeans_dataset.csv')
X = df.values  # Assuming your dataset has two columns, CGPA and IQ

# Define the number of clusters
n_clusters = 4

# Create and fit the K-Means model
kmeans = KMeans(n_clusters=n_clusters, max_iter=500, random_state=2)
y_means = kmeans.fit_predict(X)

# Plot the clusters
colors = ['red', 'blue', 'green', 'yellow']

for i in range(n_clusters):
    plt.scatter(X[y_means == i, 0], X[y_means == i, 1], color=colors[i], label=f'Cluster {i + 1}')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', label='Centroids', marker='x')
plt.legend()
plt.title(f'K-Means Clustering with {n_clusters} Clusters')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.show()
```
